## Part 4: Filtering - Using Conditionals to Filter Rows and Columns

```python
df_people
```

#### The filters return a series of true/false values and when we apply that to our data frame those true/false values basically apply a mask to our dataframes where we get all the true values back and not the false 

```python
df_people['last']=="Doe"
```

```python
filt= df_people['last']=="Doe"
```

```python
df_people[filt]
```

```python
df_people[df_people['last']=="Schafer"]
```

```python
df_people.loc[filt]
```

#### We can still grab the specific columns that we want as well. if i want the email column than i can simply add a second value

```python
df_people.loc[filt, 'email']
```

```python
filt=(df_people['last']=="Doe")& (df_people["first"]=="John")
```

```python
df_people[filt]
```

```python
df_people.loc[filt, 'email']
```

```python
filt=(df_people['last']=="Schafer")| (df_people["first"]=="John")
```

```python
df_people[filt]
```

###### We can get also the opposite of the filter for example if we want to get all the rows where the last name isn't Schafer and the first name isn't john. i can simply add a (-) at the beginning of this filter   

```python
df_people[-filt]
```

## Part 5: Updating Rows and Columns - Modifying Data Within DataFrames

#### We are going to learn how to update data within our rows and columns in our dataframes so let's go ahead and get started.

```python
df_people
```

```python
df_people.columns
```

#### the columns names have changed in the dataframe

```python
df_people
```

```python
print(pd.__version__ )
```

```python
df_people.columns=[x.upper() for x in df_people.columns]
```

```python
df_people
```

```python
df_people.columns= df_people.columns.str.replace('_', ' ')
```

```python
df_people
```

```python
df_people.columns= df_people.columns.str.replace(' ', '_')
```
